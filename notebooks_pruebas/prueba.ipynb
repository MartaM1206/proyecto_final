```python
import pandas as pd
```

```python
df_original = pd.read_csv("../datos/madrid_4_2025.csv", parse_dates = True, sep = ";")
df_original.head()
```

```python
df_pruebas = df_original[["punto_muestreo", "ano", "mes", "dia", "h01", "v01", "h02", "v02", "h03", "v03","h04", "v04"]]
df_pruebas.info()
```

```python
df_pruebas[["ano", "mes", "dia"]] = (df_pruebas[["ano", "mes", "dia"]]).astype(str)
```

```python
df_pruebas.info()
```

```python
df_pruebas["fecha"] = df_pruebas["ano"] + "-" + df_pruebas["mes"] + "-" + df_pruebas["dia"]
df_pruebas.drop(columns = ["ano", "mes", "dia"], inplace=True)
df_pruebas.info()
```

```python
df_pruebas
```

```python
df_pruebas_peque = df_pruebas[df_pruebas["punto_muestreo"].str.startswith("28102001")]
df_pruebas_peque
```

```python
# quiero probar a unir el valor de h con el valor de v para hacer el melt y luego separarlo
```

```python
# Transformar el DataFrame al formato deseado
columnas_horas = ["h01", "h02", "h03", "h04"]  # Columnas que contienen las horas
columnas_valores = ["v01", "v02", "v03", "v04"]  # Columnas que contienen los valores

# Crear un DataFrame "long" con las filas necesarias
#df_transformado = pd.DataFrame({
    #"punto_muestreo": df_pruebas_peque["punto_muestreo"].repeat(len(columnas_horas)),
    #"fecha": df_pruebas_peque["fecha"].repeat(len(columnas_horas)),
    #"hora": [hora for hora in columnas_horas for _ in range(len(df_pruebas_peque))],
    #"valor": pd.concat([df_pruebas_peque[col] for col in columnas_horas]).reset_index(drop=True)
       # })
df_horas = df_pruebas_peque.melt(
    id_vars=["punto_muestreo", "fecha"],  # Columnas que no se transforman
    value_vars=["h01", "h02", "h03", "h04"],  # Columnas de horas que queremos transformar
    var_name="hora",  # Nombre de la nueva columna que contiene las etiquetas de hora
    value_name="valor"  # Nombre de la nueva columna que contiene los valores
)
df_validado = df_pruebas_peque.melt(
    id_vars=["punto_muestreo", "fecha"],  # Columnas que no se transforman
    value_vars=["v01", "v02", "v03", "v04"],  # Columnas de valores
    var_name="val",  # Usamos las mismas etiquetas de hora para asociarlas
    value_name="validacion"  # Nueva columna para valores de mediciÃ³n
)
df_transformado = pd.merge(df_horas, df_validado, on=["punto_muestreo", "fecha", "hora"])

# Mostrar el DataFrame transformado
df_transformado
```
